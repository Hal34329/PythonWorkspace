```python
from gtts import gTTS
from IPython.display import Audio

def speak(text, language='is'):
  tts = gTTS(text=text, lang=language, slow=False)
  tts.save("output.mp3")
  return Audio("output.mp3", autoplay=True)

# Ejemplo de uso
speak("Ég er að læera íslensku.")
```

```python
x = input("Frase:")
speak(x)
```
