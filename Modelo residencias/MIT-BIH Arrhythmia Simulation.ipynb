Simulación de señal ECG

```python
import neurokit2 as nk
import matplotlib.pyplot as plt
```

Paso 2: Simular la señal ECG

Puedes utilizar la función ecg_simulate() de neurokit2 para simular una señal ECG.

```python
ecg_signal = nk.ecg_simulate(duration=10, sampling_rate=1000, heart_rate=70)
```

Paso 3: Procesar la señal ECG y obtener los picos

```python
#signals, info = nk.ecg_process(ecg_signal, sampling_rate=1000)
signals, info = nk.ecg_process(ecg_signal, sampling_rate=1000, method="pantompkins1985")
```

Paso 4: Acceder a los picos P, R, S, Q

La información de los picos se encuentra en el diccionario info. Puedes acceder a los picos R con `info['ECG_R_Peaks']`, a los picos P con `info['ECG_P_Peaks']`, a los picos Q con `info['ECG_Q_Peaks']` y a los picos S con `info['ECG_S_Peaks']`.

Paso 5: Visualizar los resultados

```python
plt.plot(ecg_signal, label="ECG")  # Señal original
plt.plot(signals['ECG_R_Peaks'], ecg_signal[signals['ECG_R_Peaks']], "ro", label="Picos R")  # Picos R
plt.plot(signals['ECG_P_Peaks'], ecg_signal[signals['ECG_P_Peaks']], "go", label="Picos P")  # Picos P
plt.plot(signals['ECG_Q_Peaks'], ecg_signal[signals['ECG_Q_Peaks']], "bo", label="Picos Q")  # Picos Q
plt.plot(signals['ECG_S_Peaks'], ecg_signal[signals['ECG_S_Peaks']], "ko", label="Picos S")  # Picos S

plt.legend()  # Mostrar leyenda
plt.show()  # Mostrar gráfico
```

```python
# Load NeuroKit and other useful packages
import neurokit2 as nk
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [15, 9]  # Bigger images
```

```python
# Simulate 30 seconds of ECG Signal (recorded at 250 samples / second)
ecg_signal = nk.ecg_simulate(duration=30, sampling_rate=360)
```

```python
# Automatically process the (raw) ECG signal
signals, info = nk.ecg_process(ecg_signal, sampling_rate=360)
```

```python
# Extract clean ECG and R-peaks location
rpeaks = info["ECG_R_Peaks"]
cleaned_ecg = signals["ECG_Clean"]
```

```python
print(rpeaks)
```

```python
r_peak_values = rpeaks['ECG_R_Peaks']
```

```python
print(r_peak_values)
```

```python
print(len(cleaned_ecg))
print(max(rpeaks['ECG_R_Peaks']))
```

```python
valid_indices = rpeaks['ECG_R_Peaks'][rpeaks['ECG_R_Peaks'] < len(cleaned_ecg)]
r_peak_amplitudes = cleaned_ecg[valid_indices]
```

```python
print(r_peak_amplitudes)
```

```python
# Visualize R-peaks in ECG signal
plot = nk.events_plot(rpeaks, cleaned_ecg)

plt.xlim(0, 5000)
plt.plot([], [], 'ro', label="R peaks")
plt.legend() 
plt.show()
```

```python
# Plotting all the heart beats
epochs = nk.ecg_segment(cleaned_ecg, rpeaks=None, sampling_rate=360, show=True)
```

```python
# Load NeuroKit and other useful packages
import neurokit2 as nk
import numpy as np
import pandas as pd
```

```python
# Retrieve ECG data from data folder
ecg_signal = nk.data(dataset="ecg_1000hz")
# Extract R-peaks locations
_, rpeaks = nk.ecg_peaks(ecg_signal, sampling_rate=1000)
```

```python
# Visualize R-peaks in ECG signal
plot = nk.events_plot(rpeaks['ECG_R_Peaks'], ecg_signal)
```

```python
# Zooming into the first 5 R-peaks
plot = nk.events_plot(rpeaks['ECG_R_Peaks'][:5], ecg_signal[:6000])
```

```python
# Delineate the ECG signal
_, waves_peak = nk.ecg_delineate(ecg_signal, rpeaks, sampling_rate=1000, method="peak")
```

```python
# Zooming into the first 3 R-peaks, with focus on T_peaks, P-peaks, Q-peaks and S-peaks
plot = nk.events_plot([waves_peak['ECG_T_Peaks'][:3], 
                       waves_peak['ECG_P_Peaks'][:3],
                       waves_peak['ECG_Q_Peaks'][:3],
                       waves_peak['ECG_S_Peaks'][:3]], ecg_signal[:4000])
```

```python
# Delineate the ECG signal and visualizing all peaks of ECG complexes
_, waves_peak = nk.ecg_delineate(ecg_signal, 
                                 rpeaks, 
                                 sampling_rate=1000, 
                                 method="peak", 
                                 show=True, 
                                 show_type='peaks')
```
