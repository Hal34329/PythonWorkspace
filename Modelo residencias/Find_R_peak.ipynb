```python
import scipy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ekg = pd.read_csv('100.csv', index_col=0)
ekg_first_4_beats = ekg[0:1000]

# Find the peaks of the R waves
peaks, metadata = scipy.signal.find_peaks(ekg_first_4_beats['MLII'], height=0.6)

# Graph the heartbeats
ekg_first_4_beats[['MLII']].plot()

# Graph the R waves as Xs on top of the heartbeats
plt.plot(peaks, ekg_first_4_beats['MLII'].iloc[peaks], 'x')
plt.title('First 4 R Waves of Patient 100\'s EKG')

plt.show()
```
