```python
#Load NeuroKit and other useful packages
import neurokit2 as nk
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [15, 9]

# Normal parameters (used by default)
# ===================================
# t, the starting position along the circle of each interval in radius
ti = np.array((-70, -15, 0, 15, 100))
# a, the amplitude of each spike
ai = np.array((1.2, -5, 30, -7.5, 0.75))
# b, the width of each spike
bi = np.array((0.25, 0.1, 0.1, 0.1, 0.4))

# Add noise
# ===============
ti = np.random.normal(ti, np.ones(5) * 3)
ai = np.random.normal(ai, np.abs(ai / 5))
bi = np.random.normal(bi, np.abs(bi / 5))

ecg_signal = nk.ecg_simulate(duration=10, method="multileads", ti=ti, ai=ai, bi=bi)
```

```python
ecg_signal_noisy = nk.signal_distort(ecg_signal["II"], noise_amplitude=0.1, noise_frequency=[50, 100])
```

```python
ecg_signal_noisy[0:10000].plot(subplots=True)
```

```python
import neurokit2 as nk
import numpy as np
import pandas as pd

# Simular segmentos con diferentes frecuencias cardacas
ecg_segment1 = nk.ecg_simulate(duration=10, sampling_rate=360, heart_rate=60, method="multileads")
#ecg_segment2 = nk.ecg_simulate(duration=5, sampling_rate=360, heart_rate=120, method="multileads")
ecg_segment3 = nk.ecg_simulate(duration=10, sampling_rate=360, heart_rate=50, method="multileads")

# Concatenar los segmentos
ecg_signal = pd.concat([ecg_segment1, ecg_segment3])
```

```python
ecg_signal[0:10000].plot(subplots=True)
```
