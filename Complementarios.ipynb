```python
# Complementario 3
# -*- coding: utf-8 -*-
# Pág. 129
print("Complementario 3")

# Un proveedor de estéreos ofrece un descuento del 10% sobre
# el precio sin IGV, de algún aparato si este cuesta $2000 o más.
# Además, independientemente de esto, ofrece un 5% de
# descuento adicional sobre el precio si la marca es "NOSY"
# Determinar cuánto pagará, con IGV incluido, un cliente
# cualquiera por la compra de su aparato. IGV = 20%

#Entradas
print("Empieza")
price = int(input("Introduzca el precio del producto:"))
mark = input("Introduzca la marca")

if price >= 2000 and mark == "NOSY":
    pd = price * 0.90
    pt = pd * 0.95

elif price >= 2000 and mark != "NOSY":
    pt = price * 0.90

elif price < 2000 and mark == "NOSY":
    pd = price * 0.95
    pt = pd * 1.2

elif price < 2000 and mark != "NOSY":
    pt = price * 1.20


#Salida
print("\n Salida")
print("USted pagará:", pt, "pesos")
```

```python
# Complementario 4
# Dada una fecha del año 2000 (representada por el día, el mes y el año
# en formato númerico dd/mm/aaa) calcule el día siguiente. Asuma que el
# mes tiene 30 días.

#Entradas
print("Ingrese la fecha:")
a = int(input("Año:"))
m = int(input("Mes:"))
d = int(input("Día:"))

#Salida
print("\nSalida:")

if d > 0 and d < 30:
    print("Mañana es:", d+1, m, a)
else:
    if m > 0 and m < 12:
        print("Mañana es:", 1, m+1, a)
    else:
        print("Mañana es:", 1, 1, a+1)
```

```python
# Complementario 9
# 98/257
# Dado un ángulo en Radianes, conviértalo a grados
# Sexagesimales y luego a Centesimales.

pi = 3.1416

# Entradas
x = int(input("Ingrese ángulo en radianes:"))

# Proceso
sxg = x * 180 / pi
cts = x * 200 / pi

# Salida
print("El valor de %d radianes en Sexagesimales es: %f°" %(x, sxg))
print("El valor de %d radianes en Centesimales es: %f°" %(x, cts))
```

```python
# Complementario 10

# Hallar la distancia entre dos puntos en el espacio.
# d = sqrt((x2-x1)^2 + (y2-y1)^2 + (z2-z1)^2)^1/2

# Entradas
print("Ingrese los valores del punto A(x1, y1, z1)")
x1 = float(input("x1:"))
y1 = float(input("y1:"))
z1 = float(input("z1:"))

print("Ingrese los valores del punto B(x2, y2, z2)")
x2 = float(input("x2:"))
y2 = float(input("y2:"))
z2 = float(input("z2:"))

# Proceso
d = ((x2-x1)**2 + (y2-y1)**2 + (z2-z1)**2)**(1/2)

# Salida
print("La distancia entre los dos puntos es %f" %(d))
```
